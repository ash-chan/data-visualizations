## 1. Maximum heart rate achieved vs Age (Scatter plot)

```python
#Drop unnecessary columns
age_max_exercise = heart_disease[["Age","Max Heart Rate","Exercise Induced Angina"]]
age_max_exercise.replace(to_replace=[0, 1], value=['No', 'Yes'])
#age_maxheartrate
```

```python
sns.lmplot(x='Age', y='Max Heart Rate', data=age_max_exercise,
          hue = 'Exercise Induced Angina',
          legend_out = True,
          palette="Set1", size = 8)
plt.title('Max Heart Rate vs Age')
```

From the scatter plot above, we note that max heart rate generally decreases as age of the patient increases. Also, patients with exercise induced angina generally have lower maximum heart rate compared to those without.

## 2. Number of patients with each type of chest pain separated by sex (Count plot)

```python
paintype_sex = heart_disease[["Chest Pain Type","Sex"]]
```

```python
paintype_sex.Sex[paintype_sex.Sex == 1] = "Male"
paintype_sex.Sex[paintype_sex.Sex == 0] = "Female"
paintype_sex
```

```python
count_plot = sns.countplot(x="Chest Pain Type", hue="Sex", data=paintype_sex)
plt.title('Number of patients with each type of Chest Pain')
print('Value 0: typical angina, Value 1: atypical angina, Value 2: non-anginal pain, Value 3: asymptomatic')
```

Bar plot of number of patients with each type of chest pain, separated by sex.
Value 0: Typical Angina, 
Value 1: Atypical Angina, 
Value 2: Non-anginal Pain, 
Value 3: Asymptomatic. 
Clearly, most patients have typical angina pain and more males suffer from chest pains across the board.


```python
bivar = heart_disease[["Resting Blood Pressure (mm Hg)","Serum Cholestrol (mg/dl)"]]
bivar
```

## 3. Bivariate density graph of Serum Cholestrol and Resting Blood Pressure (Scatter plot overlay with Density plot)

```python
plt.figure(figsize=(13, 10))
sns.set_style('whitegrid')
sns.kdeplot(bivar["Resting Blood Pressure (mm Hg)"], bivar["Serum Cholestrol (mg/dl)"])
sns.scatterplot(x="Resting Blood Pressure (mm Hg)", y="Serum Cholestrol (mg/dl)", data=bivar)
plt.ylim(100, 400)
plt.xlim(80, 200)
```

By plotting a bivariate density graph, we can see that out of all the patient data recorded, most of them are concentrated around 125 mmHg resting blood pressure and 250 mg/dl serum cholestrol.

## Data Preprocessing and Manipulation

## Data Visualization

```python
black_friday.head()
```

## 4. Distribution of purchase amounts (Histogram) 

```python
purchase_only = black_friday[["Purchase"]]
purchase_only
```

```python
plt.figure(figsize=(10, 5))
sns.distplot(purchase_only, hist = True, bins = 20)
plt.title("Distribution of purchase amounts ($)")
plt.xlabel("Purchase Amount")
plt.ylabel("Density");
```

Distribution of purchase amounts is as shown in a histogram. There are spikes in purchase amounts at $7500, $12500 and $15000.

## 5. Total purchase amount for each occupation (Bar plot)

```python
age_purchase = black_friday.loc[:, ['Occupation', 'Purchase']].groupby('Occupation').sum()
age_purchase
```

```python
plt.figure(figsize=(12,10))
sns.barplot(age_purchase.index, age_purchase['Purchase'])
plt.title("Total Purchase Amount ($) per Occupation")
plt.xlabel("Occupation")
plt.ylabel("Total purchase amount ($)");
```

We group the buyers based on their occupation and calculate the total expenditure of each occupation. The highest spenders are those from occupation 0,4 and 7.

## 6. Purchase Amount Across Different Age Groups (Box plot)

```python
city_purchase = black_friday[["Age","City_Category","Purchase"]]
city_purchase
```

```python
age_groups = ['0-17', '18-25', '26-35', '36-45', '46-50', '51-55', '55+']
mapping = {agegroup: i for i, agegroup in enumerate(age_groups)}
key = city_purchase["Age"].map(mapping)
sorted_age = city_purchase.iloc[key.argsort()]
```

```python
plt.figure(figsize=(12, 10))
sns.boxplot(x="Age", y="Purchase", data=city_purchase, hue = "City_Category", order = ['0-17', '18-25', '26-35', '36-45', '46-50', '51-55', '55+'])
plt.title("Distributions of purchase amounts for each age group")
plt.xlabel("City")
plt.ylabel("Purchase Amount")
```

Distributions of purchase amounts for each age group and within each age group, we further divide by city categories. We notice that across the board, City Category C has the highest median purchase amounts while City Category A is mostly the lowest.

# Dataset 3: Pokemon

### Background 

Data taken from: https://www.kaggle.com/abcsds/pokemon

## Data Preprocessing and Manipulation


```python
pokemon=pd.read_csv("Pokemon.csv")
pokemon
```

## 7. Distribution of total stats for each primary type (Violin plot)

```python
pokemon1 = pokemon[["Type 1","Total"]]
pokemon1
```

```python
plt.figure(figsize=(15, 10))
sns.violinplot(x='Type 1', y='Total', data=pokemon1)
```

## 8. Total stats vs Primary type split by generation (Categorical Swarm Plot) 

```python
plt.figure(figsize=(16,12))
g = sns.factorplot(x='Type 1', 
                   y='Total', 
                   data=pokemon, 
                   hue='Generation',
                   col='Generation',
                   kind='swarm', col_wrap = 3)
 
# Rotate x-axis labels
g.set_xticklabels(rotation=45)
```

We split the Pokemon based on which generation they belong in and plot the total stats vs primary type of each pokemon in a swarm plot.

## 9. Correlation between different Pokemon stats (Heatmap) 

```python
pokemon_stats = pokemon[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
pokemon_stats
```

```python
correlation = pokemon_stats.corr()
sns.heatmap(correlation, vmin=0, vmax=1, linewidths=.2)
plt.title("Correlation between different stats")
```

This heatmap maps the correlation values between different stats of pokemon. Trivially, two of the same stats (e.g. HP and HP) have perfect correlation. However, it is interesting to note that Speed and Defense have near zero correlation with one another while Defense and Sp. Defense have rather significant correlation.

## 10. Bivariate Distribution of Attack and Sp. Attack (Joint Plot with Hex Bins) 

```python
plt.figure(figsize=(15, 10))
jplot = sns.jointplot("Attack", "Sp. Atk", data=pokemon_stats, kind="hex")
```

The Attack and Sp. Attack of all the Pokemon is plotted into a joint plot with hex bins which allows us to see the individual distributions of the Attack and Sp. Attack stats across all Pokemon, as well as their bivariate distribution.
